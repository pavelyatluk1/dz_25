### ЗАВДАННЯ

Ймовірність того, що при перевірці деталь виявиться стандартною дорівнює 0,6. Перевірці підлягають 3 деталі.
Побудувати закон розподілу системи двох дискретних віпадкових величин X - появи числа бракованих деталей і Y - появи числа стандартних деталей.
Обчислити коефіцієнт кореляції r(xy).

### ВИРІШЕННЯ

1) Подія Х - поява бракованої  деталі, подія Y - поява стандартної деталі. Події X і Y є протилежними, тому ймовірність появи бракованої деталі - 

    P(X) = 1 - P(Y) = 1 - 0.6 = 0.4, де ймовірність появи стандартної деталі P(Y) = 0.6 згідно умови.
    
    
    
2) Розрахуємо ймовірності можливих варіантів перевірки деталей.
    Всого можуть бути наступні варіанти
    
    XXX, XXY, XYY, YYY
    
    
    P(XXX) = P(X) * P(X) * P(X) = 0.4 * 0.4 * 0.4 = 0.064
    P(XXY) = P(X) * P(X) * P(Y) +  P(X) * P(X) * P(Y) + P(X) * P(X) * P(Y) = 0.6 * 0.4 * 0.4 + 0.6 * 0.4 * 0.4 + 0.6 * 0.4 * 0.4 = 0.288
    P(XYY) = P(X) * P(Y) * P(Y) +  P(X) * P(Y) * P(Y) + P(X) * P(Y) * P(Y) = 0.6 * 0.6 * 0.4 + 0.6 * 0.6 * 0.4 + 0.6 * 0.6 * 0.4 = 0.432
    P(YYY) = P(Y) * P(Y) * P(Y) = 0.6 * 0.6 * 0.6 = 0.216
    
    
    
3) Згідно обчислених ймовірностей будуємо закон розподілу двох дискретних випадкових величин

| X    | Y    | 0    | 1    | 2    | 3    | P(X)  |
| :--- | :---:| :---:| :---:| :---:| :---:| ---:  |
| 0    |      | 0    | 0    | 0    | 0.216| 0.216 |            
| 1    |      | 0    | 0    | 0.432| 0    | 0.432 |            
| 2    |      | 0    | 0.288| 0    | 0    | 0.288 |            
| 3    |      | 0.064| 0    | 0    | 0    | 0.064 |
| P(Y) |      | 0.064| 0.288| 0.432| 0.216| 1     | 



3) Розрахунок математичного сподівання

    M(X) = 1 * 0.432 + 2 * 0.288 + 3 * 0.064 = 1.2
    
    M(Y) = 1 * 0.288 + 2 * 0.432 + 3 * 0.216 = 1.8
    
    M(XY) = 1 * 2 * 0.432 + 2 * 1 * 0.288 = 1.44
    
    
4) Розрахунок дисперсії

    D(X) = 1 ** 2 * 0.432 + 2 ** 2 * 0.288 + 3 ** 2 * 0.064 - 1.2 ** 2 = 0.72
    
    D(Y) = 1 ** 2 * 0.288 + 2 ** 2 * 0.432 + 3 ** 2 * 0.216 - 1.8 ** 2 = 0.72
    
    
5) Розрахунок середньоквадратичного відхилення

    sg(X) = 0.72 ** 0.5 = 0.8485
    
    sg(Y) = 0.72 ** 0.5 = 0.8485
    
    
6) Розрахунок коваріації

    cov(XY) = M(XY) - M(X) * M(Y) = 1.44 - 1.2 * 1.8 = - 0.72
    
    
7) Розрахунок коефіцієнту кореляції

    r(XY) = cov(XY) / (sg(X) * sg(Y)) = - 0.72 / (0.8485 * 0.8485) = -0.72 / 0.72 = - 1
